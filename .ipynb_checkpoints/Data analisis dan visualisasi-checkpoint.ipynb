<h2> The Type of Data </h2>
<h3> Open Data vs Private Data </h3>
<h4> Open Data:</h4>
Any content that people can free to use, reuse, and redistribute without any legal, technological, or social restriction. <br>
<h4> Private Data:</h4>
Data related to an expectation of privacy and regulated by a particular country/government.
<br>
<h3> Structured vs Unstructured Data</h3>
<h4> Structured:</h4>
<ul>
    <li>Data entered and maintained in fixed fields within a file or record</li>
    <li>Easily entered, classified, queried, and analyzed</li>
    <li>Relational databases or spreadsheets</li>
</ul>
<h4> Unstructured:</h4>
<ul>
    <li>Lacks organization</li>
    <li>Raw Data</li>
</ul>
<img src=".\images\st-vs-ust.jpg"></img>
<h3>Data at Rest vs Data at Motion</h3>
<h4> Data at Rest:</h4>
Data stored in a physical location such as a server hard drive or within a data center.
<h4> Data at Motion:</h4>
Dynamic data that requires real-time processing before the data becomes irrelevant or obsolete.<br>

<h3> Outliers</h3>
An outlier is a data-item that deviates significantly from the rest of the normal data-items.<br>
The outliers can decrease the model performance and lead to the measurement errors.


```python
sns.boxplot(x='SepalWidthCm', data=data)
```

<h3> IQR (Interquartile Range)</h3>
The quartiles are thre points that divide the data into exactly four equal parts.
<ol>
    <li>Q1 --> the middle number between the smallest number and the median of the dataset</li>
    <li>Q2 --> the median of the dataset</li>
    <li>Q3--> the middle number between the median and the highest value of the dataset</li>
</ol>

<p> The IQR tells the range where the bulk of the values lie<br>
    The IQR can be calculate by<br>
    <b><i>IQR = Q3-Q1</i></b>
</p>
<img src=".\images\iqr.png"><img>

<h4> Find the IQR of bellow dataset</h4>
1, 19, 7, 6, 5, 9, 12, 27, 18, 2, 15

<h4> Answer </h4>
Short the list:<br>
1, 2, 5, 6, 7, 9, 12, 15, 18, 19, 27 <br>
Q1: 5<br>
Q2: 9<br>
Q3: 18<br>
IQR = Q3-Q1 = 18-5 = 13<br>
<br>
lower limmit = Q1-1.5xIQR<br>
upper limmit = Q3+1.5xIQR<br>

```python
import numpy as np
# IQR
Q1 = np.percentile(data['SepalWidthCm'], 25,
                interpolation = 'midpoint')
 
Q3 = np.percentile(data['SepalWidthCm'], 75,
                interpolation = 'midpoint')
IQR = Q3 - Q1
 
print("Old Shape: ", data.shape)
 
# Upper bound
upper = np.where(data['SepalWidthCm'] >= (Q3+1.5*IQR))
 
# Lower bound
lower = np.where(data['SepalWidthCm'] <= (Q1-1.5*IQR))
 
# Removing the Outliers
data.drop(upper[0], inplace = True)
data.drop(lower[0], inplace = True)
 
print("New Shape: ", data.shape)
 
sns.boxplot(x='SepalWidthCm', data=data)
```
