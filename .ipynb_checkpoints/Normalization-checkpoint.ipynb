<h1> Normalization </h1>
<h3> What is Normalization? </h3>
<b>Normalization</b> is a scaling technique in machine learning to change the values of numeric colum into a common scale. <br>
<br>
Normalization is done during data preparation process --> preprocessing <br>
It is <b>NOT NECESSARY</b> for all datset in a model. Its only required when the range of features are different. <br><br>
<b> Example</b>: If the dataset contains fetaures "age" and "income", where age has range 0-80, and income lied between 0-100.000usd. These two features have different range. In further model the income can be influence the result more due to its large value. But it does not mean that it is more important as a predictor. 

<h3> Normalization Technique in Machine Learning </h3>

The most widely used types of normalization in machine learning:<br>
<dl>
    <dt><b>Min-Max Scaling</b></dt>
    <dd>Subtract the minimum value from each column’s highest value and divide by the range. Each new column has a minimum value of 0 and a maximum value of 1.</dd>
    <dt><b>Standardization Scaling</b></dt>
    <dd>The term “standardization” refers to the process of centering a variable at zero and standardizing the variance at one.</dd>
</dl>

<h3> When to Normalize Data? </h3>
<ul>
    <li>Normalization is used when the data does not have Gaussian distribution, or the data distribution is unknown </li>
    <li>The algorithm to trained the model does not make presumption about the data distribution, Example ANN </li> 
</ul>

<h3>When to Standardization Data?</h3>
<ul>
    <li> When the data is used for multivariate analysis</li>
    <li> When the data has gausian distribution </li>
    <li> The algorithm used make ssumptions about the data distribution, example: Logistic Regression </li>
</ul>

```python
# import required library
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib
import matplotlib.pyplot as plt
from sklearn import preprocessing

np.random.seed(0)

%matplotlib inline
matplotlib.style.use('ggplot')
```

```python
# generate data
df = pd.DataFrame({ 
    'beta': np.random.beta(5, 1, 1000) * 60,        # beta
    'normal_l': np.random.normal(10, 10, 1000),     # normal leptokurtic
})

# make bimodal distribution
first_half = np.random.normal(20, 3, 500) 
second_half = np.random.normal(-20, 3, 500) 
bimodal = np.concatenate([first_half, second_half])

df['bimodal'] = bimodal
```

```python
# plot original distribution plot
fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('Original Distributions')

sns.kdeplot(df['beta'], ax=ax1)
sns.kdeplot(df['normal_l'], ax=ax1)
sns.kdeplot(df['bimodal'], ax=ax1);
```

```python
# Min_max scaler
mm_scaler = preprocessing.MinMaxScaler()
df_mm = mm_scaler.fit_transform(df)

col_names = list(df.columns)
df_mm = pd.DataFrame(df_mm, columns=col_names)

fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('After MinMaxScaler')

sns.kdeplot(df_mm['beta'], ax=ax1)
sns.kdeplot(df_mm['normal_l'], ax=ax1)
sns.kdeplot(df_mm['bimodal'], ax=ax1)
```

```python
s_scaler = preprocessing.StandardScaler()
df_s = s_scaler.fit_transform(df)

df_s = pd.DataFrame(df_s, columns=col_names)

fig, (ax1) = plt.subplots(ncols=1, figsize=(10, 8))
ax1.set_title('After StandardScaler')

sns.kdeplot(df_s['beta'], ax=ax1)
sns.kdeplot(df_s['normal_l'], ax=ax1)
sns.kdeplot(df_s['bimodal'], ax=ax1)
```
