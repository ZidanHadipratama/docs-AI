## Pre-Processing

### Missing Values

## Clusterisasi

### KMeans

```python
df.head()
```

```python
df.info()
```

```python
# List to hold the silhouette scores
sil_scores = []

# Range of cluster values to test
cluster_range = range(2, 11)

for n_clusters in cluster_range:
    # Create an instance of KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)

    km = df.copy()

    # Fit the model to your data
    kmeans.fit(km)

    # Get cluster labels
    labels = kmeans.labels_

    # Calculate silhouette score and append to list
    score = silhouette_score(km, labels)
    sil_scores.append(score)

# Plot silhouette scores
plt.plot(cluster_range, sil_scores, marker='o')
plt.title('Silhouette scores for different numbers of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Create an instance of KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=4, random_state=0)

# Fit the model to your data and predict the cluster assignments for each data point
km['cluster'] = kmeans.fit_predict(km)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(km)

# Create a scatter plot
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=km['cluster'], cmap='viridis')

plt.title('Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```

```python
# Assuming 'kmeans' is your fitted KMeans object and 'df' is your DataFrame
centroids = kmeans.cluster_centers_

# Create a DataFrame for the centroids
# Use the columns of the DataFrame used to fit the KMeans model
centroids_df = pd.DataFrame(centroids, columns=km.columns[:centroids.shape[1]])

# Print the centroids
print(centroids_df)
```

### DBScan

```python
from sklearn.cluster import DBSCAN
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt
import numpy as np

# Assuming df is your DataFrame

# Range of epsilon values to test
eps_range = np.arange(0.1, 1.1, 0.1)

samples = range(1,10)

# Create a figure
plt.figure(figsize=(10,8))

for sample in samples:
    # List to hold the silhouette scores
    sil_scores = []

    for eps in eps_range:
        # Create an instance of DBSCAN
        dbscan = DBSCAN(eps=eps, min_samples=sample)

        km = df.copy()
    
        # Fit the model to your data
        labels = dbscan.fit_predict(km)
    
        # Calculate silhouette score and append to list
        # Note: Silhouette score is only defined when number of labels is 2 <= n_labels <= n_samples - 1
        if len(set(labels)) > 1 and len(set(labels)) < km.shape[0]:
            score = silhouette_score(km, labels)
            sil_scores.append(score)
        else:
            sil_scores.append(-1)

    # Plot silhouette scores for different epsilon values
    plt.plot(eps_range, sil_scores, marker='o', label=f'min_samples={sample}')

plt.title('Silhouette scores for different epsilon values')
plt.xlabel('Epsilon value')
plt.ylabel('Silhouette score')
plt.legend()
plt.show()
```

```python
from sklearn.decomposition import PCA

# Create an instance of KMeans with the optimal number of clusters
kmeans = DBSCAN(eps=0.2, min_samples=9) 

# Fit the model to your data and predict the cluster assignments for each data point
km['cluster'] = kmeans.fit_predict(km)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(km)

# Create a scatter plot
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=km['cluster'], cmap='viridis')

plt.title('Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```

```python
# Fit the model to your data and predict the cluster assignments for each data point
km['cluster'] = kmeans.fit_predict(km)

# Calculate the mean feature values for each cluster
cluster_means = km.groupby('cluster').mean()

# Print the mean feature values
print(cluster_means)
```

### Aglomerative Clustering

```python
from sklearn.cluster import AgglomerativeClustering
from sklearn.metrics import silhouette_score
import matplotlib.pyplot as plt

# Range of number of clusters to test
cluster_range = range(2, 11)  # Adjust this range as needed

# List to hold the silhouette scores
sil_scores = []

for n_clusters in cluster_range:
    # Create an instance of AgglomerativeClustering
    agglomerative = AgglomerativeClustering(n_clusters=n_clusters)

    km = df.copy()
    
    # Fit the model to your data
    labels = agglomerative.fit_predict(km)
    
    # Calculate silhouette score and append to list
    score = silhouette_score(km, labels)
    sil_scores.append(score)

# Plot silhouette scores for different numbers of clusters
plt.plot(cluster_range, sil_scores, marker='o')
plt.title('Silhouette scores for different numbers of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Create an instance of KMeans with the optimal number of clusters
kmeans = AgglomerativeClustering(n_clusters=4)

# Fit the model to your data and predict the cluster assignments for each data point
km['cluster'] = kmeans.fit_predict(km)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(km)

# Create a scatter plot
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=km['cluster'], cmap='viridis')

plt.title('Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```

```python
# Fit the model to your data and predict the cluster assignments for each data point
km['cluster'] = kmeans.fit_predict(km)

# Calculate the centroids
centroids = km.groupby('cluster').mean()

# Print the centroids
print(centroids)
```
