## Clusterisasi

### KMeans

```python
df.head()
```

```python
df.info()
```

```python
# List to hold the silhouette scores
sil_scores = []

# Range of cluster values to test
cluster_range = range(2, 11)

for n_clusters in cluster_range:
    # Create an instance of KMeans
    kmeans = KMeans(n_clusters=n_clusters, random_state=0)

    km = df.copy()

    # Fit the model to your data
    kmeans.fit(km)

    # Get cluster labels
    labels = kmeans.labels_

    # Calculate silhouette score and append to list
    score = silhouette_score(km, labels)
    sil_scores.append(score)

# Plot silhouette scores
plt.plot(cluster_range, sil_scores, marker='o')
plt.title('Silhouette scores for different numbers of clusters')
plt.xlabel('Number of clusters')
plt.ylabel('Silhouette score')
plt.show()
```

```python
from sklearn.decomposition import PCA

# Create an instance of KMeans with the optimal number of clusters
kmeans = KMeans(n_clusters=4, random_state=0)

# Fit the model to your data and predict the cluster assignments for each data point
km['cluster'] = kmeans.fit_predict(km)

# Perform PCA
pca = PCA(n_components=2)
pca_result = pca.fit_transform(km)

# Create a scatter plot
plt.scatter(pca_result[:, 0], pca_result[:, 1], c=km['cluster'], cmap='viridis')

plt.title('Visualization of Clusters')
plt.xlabel('Principal Component 1')
plt.ylabel('Principal Component 2')
plt.colorbar()
plt.show()
```

```python
# Assuming 'kmeans' is your fitted KMeans object and 'df' is your DataFrame
centroids = kmeans.cluster_centers_

# Create a DataFrame for the centroids
# Use the columns of the DataFrame used to fit the KMeans model
centroids_df = pd.DataFrame(centroids, columns=km.columns[:centroids.shape[1]])

# Print the centroids
print(centroids_df)
```

```python
ori['Cluster_Label'] = km['cluster']

ori.info()
```

```python
# Calculate mean values for each feature within each cluster
cluster_means = ori.groupby('Cluster_Label').mean()
print(cluster_means)
```

```python
ori.head()
```

```python
import matplotlib.pyplot as plt

# Calculate mean values for each feature within each cluster
cluster_means = ori.groupby('Cluster_Label').mean()

# Get the number of clusters
n_clusters = len(cluster_means)

# Create a grid of histograms (3 columns per row)
n_rows = (len(ori.columns[:-1]) + 2) // 3  # Calculate the number of rows
fig, axs = plt.subplots(n_rows, 3, figsize=(12, 6 * n_rows), sharey=True)

# Plot histograms for each column
for i, col in enumerate(ori.columns[:-1]):
    row, col_idx = divmod(i, 3)  # Determine the row and column index
    axs[row, col_idx].bar(cluster_means.index, cluster_means[col])
    axs[row, col_idx].set_xlabel('Cluster')
    axs[row, col_idx].set_title(f'{col}')
    axs[row, col_idx].set_xticks(range(n_clusters))
    axs[row, col_idx].grid(True)
    axs[row, col_idx].set_xlim(-0.5, n_clusters - 0.5)

plt.tight_layout()
plt.show()
```

```python
import matplotlib.pyplot as plt

# Calculate the number of items in each cluster
cluster_counts = ori['Cluster_Label'].value_counts()

# Plot histogram
plt.figure(figsize=(8, 6))
plt.bar(cluster_counts.index, cluster_counts)
plt.xlabel("Cluster")
plt.ylabel("Count")
plt.title("Peoples in Each Cluster")
plt.grid(True)
plt.show()
```
