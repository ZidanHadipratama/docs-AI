```python
import torch
import torch.nn as nn
import torch.nn.functional as F
import torch.optim as optim
import numpy as np
import random
import time
import os
```

```python
df = torch.randn(3000, 17)
```

```python
X_train, X_val, y_train, y_val = df[:2800, :-1], df[2800:, :-1], df[:2800, -1], df[2800:, -1]
```

```python
# Create random binary data for y_train and y_val
y_train = (y_train > 0).float()
y_val = (y_val > 0).float()
```

```python
from torch.utils.data import Dataset, DataLoader

class TrainDataset(Dataset):
    def __init__(self, X, y):
        self.X = X
        self.y = y
    
    def __getitem__(self, index):
        return self.X[index], self.y[index]

    def __len__(self):
        return len(self.y)
```

```python
sample_train = TrainDataset(X_train, y_train)
sample_val = TrainDataset(X_val, y_val)
```

```python
train_loader = DataLoader(dataset=sample_train, batch_size=32, shuffle=True)
val_loader = DataLoader(dataset=sample_val, batch_size=32, shuffle=False)
```

```python
class SimpleNeuralNetwork(nn.Module):
    def __init__(self, input_dim=16, output_dim=1):
        super(SimpleNeuralNetwork, self).__init__()
        self.input_dim = input_dim
        self.output_dim = output_dim
        self.nn1 = nn.Linear(input_dim, 32)
        self.nn2 = nn.Linear(32, 64)
        self.nn3 = nn.Linear(64, 16)
        self.nn4 = nn.Linear(16, output_dim)
    
    def forward(self, x):
        x = F.relu(self.nn1(x))
        x = F.relu(self.nn2(x))
        x = F.relu(self.nn3(x))
        x = F.sigmoid(self.nn4(x))
        return x
```

```python
model = SimpleNeuralNetwork(16, 1)
```

```python
train_loss_list, train_acc_list = [], []
val_loss_list, val_acc_list = [], []

def train(model, train_loader, val_loader, epochs=10, lr=0.001):
    criterion = nn.BCELoss()
    optimizer = optim.Adam(model.parameters(), lr=lr)
    for epoch in range(epochs):
        model.train()
        for i, (X_batch, y_batch) in enumerate(train_loader):
            optimizer.zero_grad()
            y_pred = model(X_batch.float())
            loss = criterion(y_pred, y_batch.view(-1, 1).float())
            loss.backward()
            optimizer.step()
            if i % 100 == 0:
                print(f"Epoch: {epoch}, Batch: {i}, Loss: {loss.item()}")
                train_loss_list.append(loss.item())
                # Print train accuracy
                y_pred = model(X_batch.float())
                y_pred = torch.round(y_pred)
                correct = (y_pred == y_batch.view(-1, 1)).sum().item()
                total = y_batch.size(0)
                print(f"Train Accuracy: {correct/total}")
                train_acc_list.append(correct/total)
        model.eval()
        with torch.no_grad():
            val_loss = 0
            for X_batch, y_batch in val_loader:
                y_pred = model(X_batch.float())
                val_loss += criterion(y_pred, y_batch.view(-1, 1).float())
            print(f"Epoch: {epoch}, Validation Loss: {val_loss.item()}")
            val_loss_list.append(val_loss.item())
            # Print val accuracy
            y_pred = model(X_val.float())
            y_pred = torch.round(y_pred)
            correct = (y_pred == y_val.view(-1, 1)).sum().item()
            total = y_val.size(0)
            print(f"Validation Accuracy: {correct/total}")
            val_acc_list.append(correct/total)
```

```python
train(model, train_loader, val_loader, epochs=100, lr=0.001)
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(10, 5))
# plt.plot(train_loss_list, label="Train Loss")
plt.plot(val_loss_list, label="Validation Loss")
# plt.plot(train_acc_list, label="Train Accuracy")
plt.plot(val_acc_list, label="Validation Accuracy")
plt.legend()
plt.show()
```
